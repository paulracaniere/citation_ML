```python
import numpy as np
import sklearn as skl
import csv
import nltk
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
node_inf_raw = pd.read_csv("./node_information.csv")
node_inf = node_inf_raw.values
all_auth = np.array([re.sub(r',Jr.?', 'Jr',
                            re.sub(r',(?=[a-z])', '',
                                   re.sub(r'\([^)]*\)?', '',
                                          re.sub(r' ', '' , a[3])))).split(",") if type(a[3]) == str else [] for a in node_inf])
print(all_auth[:10])
```

```python
authors = set()
for a,c in zip(all_auth,node_inf):
    for b in a:
        authors.add(b)

authors.remove('')
```

```python
print(list(authors)[:30])
```

```python
for a,b in zip(all_auth,node_inf):
    for c in a:
        if c != '' and len(c)<4:
            if c in authors:
                authors.remove(c)
```

```python
import networkx as nx
```

```python
graph = nx.Graph()
```

```python
graph.add_nodes_from(authors)
```

```python
for a in all_auth:
    if len(a) < 2:
        continue
    else:
        graph.add_edge(a[0], a[-1])
        for i in range(len(a) - 1):
            graph.add_edge(a[i],a[i + 1])
```
